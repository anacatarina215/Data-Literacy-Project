### World Cup Matches 

```python
df_wcm = pd.read_csv('data/WorldCupMatches.csv', sep=',')
display(df_wcm)
#we did some changes at hand like wrong spaces
```

```python
#remove the final rows with "NaN" values because they are just blanck spaces and the final game of World Cup 2014 is at index 851
index = []
for i in range(852, 4572):
    index.append(i)
    
df_wcm.drop(index, inplace=True)
display(df_wcm)
```

```python
#remove what is not important for our case
df_wcm = df_wcm.drop(columns=['Datetime', 'Stage', 'Stadium', 'City', 'Half-time Home Goals', 'Half-time Away Goals', 'Referee', 'Assistant 1', 'Assistant 2', 'RoundID', 'MatchID', 'Home Team Initials', 'Away Team Initials'])
display(df_wcm)
```

```python
#add the results and informations of the World cup 2018 and World Cup 2022 (from Wikipedia)
worldcupm1822 = pd.read_csv('data/WM_2018_2022.csv', sep=',', header=None, names=['Year','Home Team Name','Home Team Goals','Away Team Goals','Away Team Name','Win conditions','Attendance'])
display(worldcupm1822)

df_wcm = df_wcm.append(worldcupm1822, ignore_index=True)
display(df_wcm)
```

```python
#add a column with the Winner Name of each match just to be an easier way of visualize
winner = []
for row in df_wcm.iterrows():
    if row[1][2] > row[1][3]: #Home Team Goals > Away Team Goals -> Winner: Home Team
        winner.append(df_wcm['Home Team Name'][row[0]])
    if row[1][2] < row[1][3]: #Home Team Goals < Away Team Goals -> Winner: Away Team
        winner.append(df_wcm['Away Team Name'][row[0]])
    if row[1][2] == row[1][3]: #Home Team Goals = Away Team Goals -> Tie case
        if df_wcm['Win conditions'][row[0]] == " ": #means that the game ended as a tie
            winner.append('tie')
        else: #we can have the case of extratime or penalties
            split = df_wcm['Win conditions'][row[0]].split()
            for i in split:
                if i == 'penalties' or i == 'extra':
                    for j in range(0, len(split)):
                        if split[j] == 'win':
                            s = ""
                            for k in range(0,j):
                                s += " "+split[k]
                            s = s.lstrip()
                            winner.append(s)
                

print(winner)
df_wcm['Match Winner']=winner
display(df_wcm)
```

```python
#some preprocessing
df_wcm = df_wcm.replace('Germany FR', 'Germany') #due the separation before the World War II (also because of this, there is no acess to years of war - between index 2 and 3)
df_wcm = df_wcm.replace('GermanyFR', 'Germany')
df_wcm = df_wcm.replace('German DR', 'Germany')
df_wcm = df_wcm.replace('United States', 'USA')
df_wcm = df_wcm.replace('Soviet Union', 'Russia')
df_wcm = df_wcm.replace('IR Iran', 'Iran')
df_wcm = df_wcm.replace("C�te d'Ivoire", "Cote d'Ivoire")
df_wcm = df_wcm.replace("Korea Republic", "South Korea")
df_wcm = df_wcm.replace("Dutch East Indies", "Indonesia")
display(df_wcm)
```

```python
df_wcm.to_csv("data/WorldCupMatches_complete.csv")
```
